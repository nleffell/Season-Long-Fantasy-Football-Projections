```python
import pandas as pd
import numpy as np
import requests as re
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split 
```

```python
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
def get_stats(year):
    r = re.get(f'https://www.pro-football-reference.com/years/{year}/fantasy.htm')
    soup = BeautifulSoup(r.content, 'html.parser')
    table = soup.find_all('table')[0]  
    df = pd.read_html(str(table))[0]
    
    #drop top level column names (there are 2 rows of columns headers)
    df.columns = df.columns.droplevel(0)
    
    #remove bogus entires where Player is equal to 'Player'
    df = df[df['Player'] != 'Player'] 
    
    #replace nuls with 0s
    df.fillna(0, inplace=True)
    
    #correct data types
    for col in df.columns:
        if col not in ['Tm', 'Player', 'FantPos']:
            if col not in ['Y/A', 'Y/R', 'PPR', 'DKPt', 'FDPt']:
                df[col] = df[col].astype(int)     
            else:
                df[col] = df[col].astype(float)
    
    #fix columns - add more specific column names i.e. differeniate between passing, rushing, recieving yards and TDs
    stat_details = {8:'Pass_Att', 9:'Pass_Yds', 10:'Pass_TD', 12:'Rush_Att', 13:'Rush_Yds',
                14:'Rush_Y/A', 15:'Rush_TD', 18:'Rec_Yds', 19:'Rec_Y/R', 20:'Rec_TD'}
    df.columns = [stat_details[x[0]] if x[0] in stat_details.keys() else x[1] for x in enumerate(df.columns.values)]
    
    #remove redundant columns
    df = df.drop(columns=['FantPt', 'DKPt', 'FDPt', 'PosRank', 'OvRank', 'TD'])
    
    #remove non letter characters from names, add the year we are predicting for (so year + 1) to differeniate between
    # players that appear more than once in this dataset
    df['Player'] = df['Player'].map(lambda x: ''.join(i for i in x if i.isalnum() or i == ' ')) + f' {year + 1}'
        
    #take only top 300 fantasy players
    df = df.sort_values(by='PPR', ascending=False)
    df = df[:300]
    
    #remove any duplicates, there were 3 extra rows when we concatenated, luckily no one with the same name is in the top 300
    df.drop_duplicates(subset=['Player'], inplace=True)
    
    #rams and chargers team names were change in 2016 and 2017 respectively, making the adjustment so its always the same
    df['Tm'] = df['Tm'].map(lambda x: 'LAR' if x == 'STL' else x)
    df['Tm'] = df['Tm'].map(lambda x: 'LAC' if x == 'SDG' else x)
    
    return df
```

```python
#add final fantasy point numbers
def get_fantasy_points(year):
    r = re.get(f'https://www.pro-football-reference.com/years/{year}/fantasy.htm')
    soup = BeautifulSoup(r.content, 'html.parser')
    table = soup.find_all('table')[0]  
    df = pd.read_html(str(table))[0]
    
    #drop top level column names (there are 2 rows of columns headers)
    df.columns = df.columns.droplevel(0)
    
    #remove bogus entires where Player is equal to 'Player'
    df = df[df['Player'] != 'Player'] 
    
    #replace nuls with 0s
    df.fillna(0, inplace=True)
    
    #only want player and fantasy points column
    df = df[['Player', 'Tm', 'PPR']]
    df['Player'] = df['Player'].map(lambda x: ''.join(i for i in x if i.isalnum() or i == ' ')) + f' {year}'
    df['PPR'] = df['PPR'].astype(float)
    df.rename(columns={'PPR':'Final_PPR_Score', 'Tm':'New_Team'}, inplace=True)
    
    #rams and chargers team names were change in 2016 and 2017 respectively, making the adjustment so its always the same
    df['New_Team'] = df['New_Team'].map(lambda x: 'LAR' if x == 'STL' else x)
    df['New_Team'] = df['New_Team'].map(lambda x: 'LAC' if x == 'SDG' else x)

    #remove any duplicates, there were 3 extra rows when we concatenated, luckily no one with the same name is in the top 300
    df.drop_duplicates(subset=['Player'], inplace=True)
    
    return df
```

### Add features

```python
#add coaching change columns, meaning they had a new coach the following year
#data from https://footballscoop.com/news/2015-16-head-coaching-changes
new_coaches_dict = {2014:['CLE', 'DET', 'HOU', 'MIN', 'TAM', 'TEN', 'WAS'], 2015:['ATL', 'BUF', 'CHI', 'DEN', 'NYJ', 'OAK', 'SFO'],
                    2016:['CLE', 'MIA', 'NYG', 'PHI', 'SFO', 'TAM', 'TEN'], 2017:['BUF', 'DEN', 'LAR', 'JAX', 'LAC', 'SFO'],
                    2018:['ARI', 'CHI', 'DET', 'IND', 'NYG', 'TEN', 'OAK'], 2019:['ARI', 'CLE', 'CIN', 'DEN', 'GNB', 'MIA', 'NYJ', 'TAM']}
```

```python
#add coordinator change columns, meaning they had a new coordinator the following year
#data from https://www.profootballrumors.com/
new_coordinators_dict = {2014:['BAL', 'CIN', 'CLE', 'DET', 'HOU', 'MIA', 'MIN', 'NYG', 'LAC', 'TAM', 'TEN', 'WAS'],
                         2015:['ATL', 'BAL', 'BUF', 'CHI', 'CLE', 'DAL', 'DEN', 'GNB', 'HOU', 'JAX', 'NYJ', 'OAK', 'SFO', 'LAR', 'TAM'],
                         2016:['CHI', 'CLE', 'CIN', 'KAN', 'MIA', 'PHI', 'LAC', 'SFO', 'TAM', 'TEN'],
                         2017:['ATL', 'BUF', 'DEN', 'HOU', 'JAX', 'LAR', 'MIN', 'NYJ', 'OAK', 'WAS'],
                         2018:['ARI', 'BUF', 'CAR', 'CHI', 'CIN', 'CLE', 'DEN', 'GNB', 'IND', 'KAN', 'LAR', 'MIA', 'MIN', 'NYG', 'NYJ', 'OAK', 'PHI', 'PIT', 'SEA', 'TEN'],
                         2019:['ARI', 'ATL' , 'BAL', 'CIN', 'CLE', 'DAL', 'DEN', 'DET', 'GNB', 'HOU', 'JAX', 'MIA', 'MIN', 'NYJ', 'TAM', 'TEN', 'WAS']}
```

```python
def combine_data_and_add_coach_features(beg_year, end_year):
    stats_df_list = []
    fp_df_list = []
    for year in range(beg_year, end_year):
        temp_df = get_stats(year)
        temp_df['New_Coach'] = temp_df['Tm'].map(lambda x: 1 if x in new_coaches_dict[year + 1] else 0)
        temp_df['New_Coordinator']  = temp_df['Tm'].map(lambda x: 1 if x in new_coordinators_dict[year + 1] else 0)
        stats_df_list.append(temp_df)
        
        fp_df_list.append(get_fantasy_points(year + 1))
    
    stats_df = pd.concat(stats_df_list) #will be my features
    fp_df = pd.concat(fp_df_list) #this is my target column
        
    df = stats_df.merge(fp_df, on='Player', how='left')
    
    #drop NAN values - players that either retired or weren't in the league the following season
    df.dropna(inplace=True)
        
    return df
```

```python
df = combine_data_and_add_coach_features(2013, 2019)
```

### Basic Linear Regression Model

```python
#load model ready data
df = pd.read_csv('model_ready_data.csv')
df = df.set_index('Player')
```

```python
X = df.drop(columns='Final_PPR_Score')
X = pd.get_dummies(X, columns=['FantPos'], drop_first=True)
y = df['Final_PPR_Score']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=7)
model = LinearRegression()
model.fit(X_train, y_train)
model.score(X_train, y_train), model.score(X_test, y_test)
```

**50.4% of fantasy points variance is explained by this lr model with data it has not yet seen**

```python
pd.DataFrame(list(zip(X.columns, model.coef_)), columns=['Feature', 'Coefficient'])
```

**Fantasy points are expected to decrease by 2.98 for each 1 unit increase in age, holding all else constant.**

**For every 1 unit increase in a players preseason fantasy ranking (worse ranking) their fantasy points are expected to decrease by .398 points**

**A player with a new coordinator but not new coach, was expected to see an increase of 8.17 fantasy points, holding all else constant**

**Coefficients for the positions of RB, WR, and TE were -74, -68, and -71, respectively when compared to the baseline position (QB). That makes sense because QBs are the highest scoring position in fantasy football.**

```python
preds = model.predict(X_test)
plt.figure(figsize=(12, 8))
plt.scatter(y_test, preds)
plt.title('True points vs predicted points', size=18)
plt.xlabel('Actaul Points Scored', size=14)
plt.ylabel('Predicted Points Scored', size=14);
```

**For a simple linear regression, this plot looks pretty good. There is a very clear line there but it definitely has room for improvement**

```python
metrics.mean_squared_error(y_test, model.predict(X_test), squared=False)
```

**RMSE of 62.33 is not great considering the smaller size of the data. That means that our predicitons are off by 62.33 fantasy points on average. I want to use RMSE instead of MAE because RMSE gives a relatively high weight to large errors which are particularly undesirable in this case.**

```python
mean_list = []
for x in range(len(y_test)):
    mean_list.append(y.mean())
metrics.mean_squared_error(y_test, mean_list, squared=False)
```

**Just for reference, the baseline of just predicting the mean every time would have an RMSE of 88.61**
